 # Importing the libraries

```python
import numpy as np 
import pandas as pd
```

 # ***Importing dataset***

```python
dataset=pd.read_csv(r"D:\Downloads\Fish.csv")
dataset.head()
```

# Data Description

### Species: Species name of fish
### Weight: Weight of fish in gram
### Length1: Vertical length in cm
### Length2: Diagonal length in cm
### Length3: Cross length in cm
### Height: Height in cm
### Width: Diagonal width in cm

#### The dataset contains 159 rows and 7 columns, where 6 columns are Numeric and 1 is Category.

# Index slicing using iloc

```python
x = dataset.iloc[:,[2,3,4,5,6]].values
y = dataset.iloc[:, 1].values


```

```python
print(y)
y.shape
```

```python
print(x)
x.shape
```

 # **Splitting the dataset into the Training set and Test set**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
x_train
```

#  ***Training the Multiple Linear Regression model on the Training set***

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

#  ***Predicting the Test set results***

```python
y_pred=regressor.predict(x_test)
y_pred
```

#  Differance between Actual weight and Prediction Weight

```python
diff=y_test-y_pred
diff
```

```python
pd.DataFrame(np.c_[y_test, y_pred, diff],columns=["Actual_value", "Predicted_value","Diff"])
```

##  ***Checking Train and Test accuracy Score***

```python
print("Train accuracy :  ",regressor.score(x_train,y_train))
print("Test accuracy :  ",regressor.score(x_test,y_pred))
```
